```python
for cont in cont_var:
    data[cont] = np.log(data[cont]+1)

data[cont_var]
```

```python
for cat in cat_feat:
  temp = data.groupby(cat)['Attrition'].count()/len(data)
  temp_df = temp[temp>0.01].index
  data[cat] = np.where(data[cat].isin(temp_df),data[cat],'Rare_var')
```

```python
data.head(20)
```

```python
for feature in cat_feat:
    labels_ordered=data.groupby([feature])['Attrition'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    data[feature]=data[feature].map(labels_ordered)
```

```python
data
```

```python
feature = [feat for feat in data.columns if feat not in ['id','Attrition']]

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(copy=True,feature_range=(0,1))
scaler.fit(data[feature])
```

```python
new_train = pd.concat([data[['id','Attrition']].reset_index(drop=True),pd.DataFrame(scaler.transform(data[feature]),columns = feature)],axis=1)
```

```python
new_train.to_csv('Xtrain.csv',index=False)
```
